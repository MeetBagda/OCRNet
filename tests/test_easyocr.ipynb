# EasyOCR Test Notebook

This notebook tests EasyOCR on 3-5 sample images from the dataset and visualizes the OCR results.

## 1. Import Required Libraries
Import EasyOCR, matplotlib, os, and glob for OCR and image handling.

```python
import easyocr
import matplotlib.pyplot as plt
import os
from glob import glob
```

## 2. Set Up EasyOCR Reader
Initialize the EasyOCR reader for English language and specify GPU usage.

```python
reader = easyocr.Reader(['en'], gpu=False)
```

## 3. Load Image Paths
Set the path to the image directory and use glob to collect image file paths.

```python
image_dir = os.path.join('..', 'dataset', 'ocr_dataset', 'train_val_images', 'train_images')
image_paths = glob(os.path.join(image_dir, '*.jpg'))
print(f"Found {len(image_paths)} images.")
for i, path in enumerate(image_paths[:5]):
    print(f"Image {i+1}: {path}")
```

## 4. Select Test Images
Select a subset (first 5) of images for OCR testing.

```python
test_images = image_paths[:5]
print(f"Testing {len(test_images)} images.")
```

## 5. Run OCR on Test Images
Use EasyOCR to extract text and confidence scores from each test image.

```python
for idx, img_path in enumerate(test_images):
    print(f'\nTesting image {idx+1}: {img_path}')
    results = reader.readtext(img_path)
    for bbox, text, conf in results:
        print(f'Text: {text}, Confidence: {conf:.2f}')
```

## 6. Display Images with OCR Results
Visualize each test image using matplotlib and optionally annotate with OCR results.

```python
for idx, img_path in enumerate(test_images):
    img = plt.imread(img_path)
    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.title(f'OCR Results for Image {idx+1}')
    plt.axis('off')
    plt.show()
    # Show detected text
    results = reader.readtext(img_path)
    print(f"Detected text in image {idx+1}:")
    for bbox, text, conf in results:
        print(f"- {text} (Confidence: {conf:.2f})")
```
